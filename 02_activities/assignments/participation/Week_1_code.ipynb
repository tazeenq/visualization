- Three qualities of data visualization: Aesthetic (is it nice to look at), substantive (does it accurately and honestly represent data), and perceptual (what message is the maker trying to convey with the visualization).

- When designing data visualizations, focus on: Context (where and how it will be used, sources like magazines vs. online), Audience (who is intended to use the vis.), and Data structure (what information does our data capture, quantities, relationships).

- How does matplotlib work? 
  - It treats a figure like a container holding a set of axes. 
  - The 'axes' are actual plot or graph [This axes is not the same as 'x' or 'y' axes, the full plot is called an axes]. 
  - A figure can hold multiple axes (multiple little plots and subplots). 
  - Every visual element of the plot - color, legend, axis title, scales, and text - is called an 'artist' and belongs to a particular axes (not the figure).


```python
# Live coding with Ciara
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
# Sample data for a figure
np.random.seed(613) # setting a random seed to make it reproducibile, could be any number.
x = np.arange(50) # x picks a random number from a range of 50 values from 0 to 49.
y = np.random.randint(0, 100, 50) # y generates random numbers from 0 (minimum) to 100 (maximum) and we need 50 values generated in the range (0-100). Syntax: 0, 100 is minimum and maximum range, 50 is number of values I need.
```

```python
# A basic scatter plot
fig, ax = plt.subplots(figsize = (5, 3)) # We are saying create a fig = figure/container, and put in a plot/graph ax = axes. Size is inches by default in matplotlib, we want 5 in by 3 in plot. 
#Here, 5 by 3 size is for both container and axes. We can put in multiple plots/graphs in one container/figure, the sizing there gets complicated.
ax.scatter(x, y) # axes should be a scatter plot using x and y variables we've made in the previous code block.
```

```python
# A bar chart instead of scatter plot
fig, ax = plt.subplots(figsize = (5, 3))
ax.bar(x, y) # using the exact same values here but I created a bar chart = bar
```

```python
# A horizontal bar chart
fig, ax = plt.subplots(figsize = (5, 3))
ax.barh(x, y) # axes should be a horizontal bar chart = barh
```

```python
# A line chart
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y) # axes should be a line chart = plot
```

```python
# A histogram
fig, ax = plt.subplots(figsize = (5, 3))
#ax.hist(x, y) -- this will throw an error because histogram takes only one variable as an argument (just counts the number of occurrences of each value; there are five 40s, seven 50s, and so on), solution is
ax.hist(y) # is I use 'x' as an argument, the histogram is just one block (there is one occurence of each value; like 0, 1, 2, 3, 4...).
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y)

# Adding axis labels and titles
ax.set_title('Total growth over time') # this is the chart title.
ax.set_ylabel('Total growth') # this is y-axis title.
ax.set_xlabel('Years since start') # this is x-axis title.
fig.tight_layout() # tight layout means 
```

```python
# Modifying labels and titles - using font dictionaries (fontdict)
font1 = {'family':'sans-serif', 'color':'blue', 'size':20} # populated with key-value pairs
font2 = {'family':'monospace', 'color':'green', 'size':14}
# Check what fonts are available in Python documentation

# Making our line plot
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y)

# Setting titles AND defining fonts
ax.set_title('Total growth over time', fontdict = font1) # title should adhere to font1 dictionary
ax.set_ylabel('Total growth', fontdict = font2) 
ax.set_xlabel('Years since start', fontdict = font2) 
fig.tight_layout() 
```

```python
# Making our line plot
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y)

# Moving the locations/alignment of labels and titles
ax.set_title('Total growth over time', fontdict = font1, loc = 'left') # title should be left aligned to the axes/plot.
ax.set_ylabel('Total growth', fontdict = font2) 
ax.set_xlabel('Years since start', fontdict = font2) 
fig.tight_layout() 
```

```python
# Changing shape and color of data markers
fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, 
           y,
           marker = '*', # Other options: > arrows, o for circles, + plus sign - more marker options in Python documentation.
           color = 'red') # single or double apostrophe will do.
fig.show()
```

```python
# Changing the style and width of lines connecting data points
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, 
           y,
           marker = '>',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2) 
```

```python
# Changing the color of data points
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, 
       y, 
       marker = '*',
       color = '#7425b9', # We can use hex codes instead of color names. Allows greater customizability.
       linestyle = '--',
       linewidth = 2)
```

```python
# Changing the color of marker body and outline, and size
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, 
       y, 
       marker = '*',
       markersize = 12,
       color = '#7425b9', 
       linestyle = '--',
       linewidth = 2,
       markeredgecolor = '#fa9359',
       markerfacecolor = '#000000'
       )
```

```python
# Adding gridlines to a graph
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, 
       y, 
       marker = '*',
       markersize = 12,
       color = '#7425b9', 
       linestyle = '--',
       linewidth = 2,
       markeredgecolor = '#fa9359',
       markerfacecolor = '#000000'
       )
ax.grid(axis = 'x', color = 'gray', linewidth = 2, linestyle = '-.')
```
